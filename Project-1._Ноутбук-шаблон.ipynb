3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Создадим вспомогательную функцию для выделения признака 'Опыт работы(месяц)'
def experience(arg):
    # Произведем проверку является ли объект NaN  или 'Не указано'
    # Если является то возвращаем NaN, если нет идем дальше
    if arg is np.nan or arg == 'Не указано':
        return np.nan
    # создаем проверочные списки для проверки причастности числа к году или к месяцу
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    # выделяем необходимый для работы участок текста
    arg_splitted = arg.split(' ')[2:7]
    # задаем начальное количество месяцев и лет (равное нулю)
    years = 0
    months = 0
    # создаем цикл, для вычисления суммарного опыта работы в месяцах
    for index, item in enumerate (arg_splitted):
        if item in year_words:
            years = int(arg_splitted[index-1])
        if item in month_words:
            months = int(arg_splitted[index-1])
    # возвращаем полученный результат
    return int(years*12 + months)

# применим созданную функцию, для заполнения столбца 'Опыт работы (месяц)'
data_dt['Опыт работы (месяц)'] = data_dt['Опыт работы'].apply(experience)

# найдем медианное значение опыта работы
print(round(data_dt['Опыт работы (месяц)'].median()))

# удалим столбец 'Опыт работы'
data_dt.drop('Опыт работы', axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Создадим вспомогательную функцию для удаления данных о близости метро
def metro(inf):
    # разделяем данные по признаку ' , '
    inf_splited = inf.split(' , ')
    # задаем условие для удаления данных о метро 
    # (если данная информация указывается в данной таблице, она указана второй, 
    # при этом общее количество элементов данного списка 4 и более)
    if len (inf_splited) >= 4:
        inf_splited.pop(1)
    return inf_splited

# применим созданную функцию, для корректировки столбца 'Город, переезд, командировки' 
data_dt['Город, переезд, командировки'] = data_dt['Город, переезд, командировки'].apply(metro)
# Создадим вспомогательную функцию для выделения признака 'Город'
def city(inf):
    # зададим список городов-миллионников
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    # задаем проверочные условия согласно задания (первый элемент в списке является городом)
    if inf[0] == 'Москва':
        return 'Москва'
    elif inf[0] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif inf[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
# применим созданную функцию, для заполнения столбца 'Город'
data_dt['Город'] = data_dt['Город, переезд, командировки'].apply(city)

# с помощью функции value_counts найдем процентное соотношение соискателей в выбранных категориях     
display (round(data_dt['Город'].value_counts(normalize=True)*100))
```

```python
# Создадим вспомогательную функцию для выделения признака 'Готовность к переезду'
def moving(inf):
    # выделим необходимый участок текста (информация о готовности к переезду хранится во 2м элементе списка)
    mov_split = inf[1].split(' ')
    # задаем условие для проверки на предмет готовности к переездам
    for i in mov_split:
        if i == 'не':
            return False
        else:
            return True
    return (mov_split)
    
# применим созданную функцию, для заполнения столбца 'Готовность к переезду'
data_dt['Готовность к переезду'] = data_dt['Город, переезд, командировки'].apply(moving)   

# Создадим вспомогательную функцию для выделения признака 'Готовность к командировкам'
def trip(inf):
    # Фильтруем список с данными. Если длина списка больше 2, значит информация о командировке задана, 
    # значит задаем проверочное условие о готовности к командировкам. 
    # Если длина списка 2 и менее, значит информация о командировках не задана, значит согласно условий задания ставим False
    if len(inf) > 2:
        trip_split = inf[2].split(' ')
        # задаем условие для проверки на предмет готовности к переездам
        for i in trip_split:
            if i == 'не':
                return False
            else:
                return True
    else:
        return False

# применим созданную функцию, для заполнения столбца 'Готовность к переезду'
data_dt['Готовность к командировкам'] = data_dt['Город, переезд, командировки'].apply(trip)   

# Произведем расчет процентного количества кандидатов готовых и к переезду и к командировкам
# создадим фильтрующую маску готовых к переезду
mask1 = data_dt['Готовность к переезду'] == True
# создадим фильтрующую маску готовых к командировкам
mask2 = data_dt['Готовность к командировкам'] == True
# применяем созданные маски
a = data_dt[mask1 & mask2]
# рассчитываем процентное количество кандидатов готовых и к переезду и к командировкам
print (round((a.shape[0] * 100)/ data_dt.shape[0]))

# удалим столбец 'Город, переезд, командировки'
data_dt.drop('Город, переезд, командировки', axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# напишем вспомогательную функцию для разделения данных в столбце 'Занятость'
def employment(emp):
    # разделяем данные по признаку ', '
    emp_splited = emp.split(', ')
    return emp_splited
# преобразуем столбец 'Занятость'
data_dt['Занятость_1'] = data_dt['Занятость'].apply(employment)
# создадим дополнительные столбцы, содержащие информацию о наличии категории занятости в списке указанном соискателем
data_dt['полная занятость'] = data_dt['Занятость_1'].apply(lambda x: True if 'полная занятость' in x else False)
data_dt['частичная занятость'] = data_dt['Занятость_1'].apply(lambda x: True if 'частичная занятость' in x else False)
data_dt['проектная работа'] = data_dt['Занятость_1'].apply(lambda x: True if 'проектная работа' in x else False)
data_dt['волонтерство'] = data_dt['Занятость_1'].apply(lambda x: True if 'волонтерство' in x else False)
data_dt['стажировка'] = data_dt['Занятость_1'].apply(lambda x: True if 'стажировка' in x else False)

# напишем вспомогательную функцию для разделения данных в столбце 'График'
def shedule(she):
    # разделяем данные по признаку ', '
    she_splited = she.split(', ')
    return she_splited
# преобразуем столбец 'График'
data_dt['График_1'] = data_dt['График'].apply(shedule)
# создадим дополнительные столбцы, содержащие информацию о наличии категории графика в списке указанном соискателем
data_dt['полный день'] = data_dt['График_1'].apply(lambda x: True if 'полный день' in x else False)
data_dt['сменный график'] = data_dt['График_1'].apply(lambda x: True if 'сменный график' in x else False)
data_dt['гибкий график'] = data_dt['График_1'].apply(lambda x: True if 'гибкий график' in x else False)
data_dt['удаленная работа'] = data_dt['График_1'].apply(lambda x: True if 'удаленная работа' in x else False)
data_dt['вахтовый метод'] = data_dt['График_1'].apply(lambda x: True if 'вахтовый метод' in x else False)

# найдем количество людей, которые ищут проектную работу и волонтёрство
# создадим фильтрующую маску людей ищущих проектную работу
mask3 = data_dt['проектная работа'] == True
# создадим фильтрующую маску ищущих волонтерство
mask4 = data_dt['волонтерство'] == True
# применяем созданные маски
b = data_dt[mask3 & mask4]
# выведем количество людей, которые ищут проектную работу и волонтёрство
print(b.shape[0])

# найдем количество людей, которые хотят работать вахтовым методом и с гибким графиком
# создадим фильтрующую маску людей ищущих работу вахтовым методом
mask5 = data_dt['вахтовый метод'] == True
# создадим фильтрующую маску работу с гибким графиком
mask6 = data_dt['гибкий график'] == True
# применяем созданные маски
с = data_dt[mask5 & mask6]
# выведем количество людей, которые хотят работать вахтовым методом и с гибким графиком
print(с.shape[0])

# удалим ненужные столбцы
data_dt.drop('Занятость', axis=1, inplace=True)
data_dt.drop('Занятость_1', axis=1, inplace=True)
data_dt.drop('График', axis=1, inplace=True)
data_dt.drop('График_1', axis=1, inplace=True)
```

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Создадим столбец с датой публикации в формате DateTime
data_dt['date'] = pd.to_datetime(data_dt['Обновление резюме'], dayfirst=True).dt.date
data_dt['date'] = pd.to_datetime(data_dt['date'])

# Выгрузим новый DF с курсами валют
data_rate = pd.read_csv('data/ExchangeRates.csv', sep=',')
# Преобразуем столбец 'date' в формат DateTime
data_rate['date'] = pd.to_datetime(data_rate['date'], dayfirst=True)

# удаляем ненужные столбцы из data_rate
data_rate.drop('per', axis=1, inplace=True)
data_rate.drop('vol', axis=1, inplace=True)
data_rate.drop('time', axis=1, inplace=True)
# создадим вспомогательную функцию для выделения желаемой ЗП из столбца 'ЗП'
def money(sum):
    # разделяем данные по признаку ' '
    mon_splited = sum.split(' ')
    return float(mon_splited[0])

# применяем полученную функцию, создаем новый столбец 'Желаемая ЗП'
data_dt['Желаемая ЗП'] = data_dt['ЗП'].apply(money)

# создадим вспомогательную функцию для выделения указанной валюты из столбца 'ЗП'
def rate(mon):
    # разделяем данные по признаку ' '
    mon_splited = mon.split(' ')
    return mon_splited[1]
# применяем полученную фуекцию, создаем новый столбец 'Валюта'
data_dt['currency'] = data_dt['ЗП'].apply(rate)

# создадим вспомогательную функцию для приведения указанных валют к стандарту ISO
def change_rate(rate):
    if rate == 'руб.':
        return 'RUB'
    elif rate =='бел.руб.':
        return 'BYN'
    elif rate == 'грн.':
        return 'UAH'
    elif rate == 'грн.':
        return 'UAH'
    elif rate == 'сум':
        return 'UZS'
    else:
        return rate
    
# Применяем функцию. 
data_dt['currency'] = data_dt['currency'].apply(change_rate)

# объединим две таблицы методом merge  по признакам 'date' b 'currency'
merge_data = data_dt.merge(    
    data_rate,
    how= 'left',
    on=('date','currency')
)

# Произведем замену NaN в столбцах 'close' и 'proportion' на 1
merge_data['close'] = merge_data['close'].fillna(1)
merge_data['proportion'] = merge_data['proportion'].fillna(1)
# Создадим новый столбец 'ЗП(руб)' и заполним его желаемой ЗП, переведенной в рубли
data_dt['ЗП(руб)'] = (merge_data['Желаемая ЗП'] * merge_data['close']) / merge_data['proportion']

# рассчитаем медианную зарплату в рублях
print (data_dt['ЗП(руб)'].median())

# удалим лишние столбцы
data_dt.drop('ЗП', axis=1, inplace=True)
data_dt.drop('Желаемая ЗП', axis=1, inplace=True)
data_dt.drop('currency', axis=1, inplace=True)
data_dt.drop('Обновление резюме', axis=1, inplace=True)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
log_age = np.log(drop_data['Возраст']+1)
histplot = sns.histplot(log_age, bins=30, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 4*log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3*log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Логарифмичное распределение возраста соискателей');

# введем функцию вычтисления выбросов по методу 3х сигм
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
# применим функцию для столбца "Возраст"
outliers, cleaned_data = outliers_z_score_mod(drop_data, 'Возраст', left=3,  right=4, log_scale=True)
# выведем количество рассчитаных выбросов
print(outliers.shape[0])
# выведем таблицу с рассчитаными выбросами
display(outliers)
```
